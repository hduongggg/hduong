```python
#W4A6
import math as mt
for _ in range(1000):
    try:
        a, b = map(int,input("Nhập hai số nguyên dương:").split())
        if a <= 0 or b <= 0 or a > b:
            print("Lỗi. Vui lòng nhập lại!")
            continue
        break
    except ValueError:
        print("Lỗi. Vui lòng nhập lại!")

total = 0
for i in range (a,b+1):
    if i < 2:
        continue
    flag = True    
    for t in range(2, int(mt.sqrt(i)) + 1):
        if i % t ==0:
           flag = False
           break
    if flag == True:
        total += i
print(f"Tổng các số nguyên tố trong đoạn [{a},{b}] là {total}")
```

```python
#W4A8
while True:
    try:
        n = int(input("Nhập số nguyên dương:"))
        if n <= 0:
            print("Vui lòng nhập số nguyên dương!")
            continue
        break
    except ValueError:
        print("Sai số liệu. Vui lòng nhập lại!")
original_n = n
for step in range(1000):
    reverse_n = []
    for i in str(n):
        reverse_n = [i] + reverse_n
    rn = int("".join(reverse_n))
    if n == rn:
        print("Số bước là ", step + 1)
        print(f"Giá trị palindrome của số nguyên dương {original_n} là {rn}")
        break
    n = n + rn
else:
    print("Không tìm được palindrome sau 1000 bước thử cho số ", original_n)
```

```python
###### W4A10
for _ in range(1000):
    try:
        n = int(input("Nhập số nguyên dương:"))
        if n <= 0:
            print("Vui lòng nhập số nguyên dương!")
            continue
        break
    except ValueError:
        print("Sai số liệu. Vui lòng nhập lại!")
max_L = 0
L_x = 1
for x in range(1, n+1):
    k = x
    L = 1
    for _ in range(100000):
        if k == 1:
            break
        if k % 2 == 0:
            k //= 2
        else: 
            k = 3 * k + 1
        L += 1
    else: 
        print("Chuỗi không hội tụ sau 100000 bước chạy")
    if L > max_L or (L == max_L and x < L_x):
        max_L = L
        L_x = x

print(f"Số có độ dài Collatz lớn nhất là {L_x} với độ dài {max_L}")
    
```

```python
#W4A14
#Cach 2 #Euclid
m, n = map(int,input("Nhập hai số nguyên dương:").split())
for _ in range(min(m, n)):
    if n == 0:
        break
    print(m, n)
    m, n = n, m % n
    print(m,n)

print("Ước chung lớn nhất là",m)  
```
