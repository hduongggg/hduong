```python
#Bai 4
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    print(" " * (h - i), end="")
    if i == 1:
        print("*")
    elif i == h:
        print("*" * (2 * i - 1))
    else:
        print("*" + " " * (2 * i - 3) + "*")
    

    
```

```python
#Bai 9
import math as mt
a = int(input("Nhập vào số nguyên dương:"))
if mt.sqrt(a) % 1 == 0:
    print(a, "là số chính phương")
else:
    print(a,"không là số chính phương")
    
```

```python
#Bai 13
A = int(input("Nhập số nguyên dương:"))
total = 0
for n in range(1,1000000):
    total += 1 / n
    if total > A:
        print("Giá trị n nhỏ nhất là:", n)
        break
else:
    print("Không tìm thấy giá trị n phù hợp sau 1000000 bước thử")

    
    
```

```python
#Bai 14
flag = True
for i in range(10000):
    try:
        x = int(input(f"Nhập số nguyên thứ {i + 1}: "))
    except ValueError:
        print("Nhập lại số nguyên!")
        continue
    if x == -1:
        break
    if flag == True:
        max_x = x
        min_x = x
        flag = False
    else:
        if x > max_x:
            max_x = x
        if x < min_x: 
            min_x = x
if flag == True: 
    print("Dữ liệu nhập vào không hợp lệ!")
else:
    print(f"Số lớn nhất: {max_x}\nSố nhỏ nhất: {min_x}")
    
    
    
    
```
