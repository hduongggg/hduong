```python
#W6A1
def bai1(n):
    A = []
    for i in n:
        if i not in A:
            A.append(i)
    return A
n = list(map(int,input().split()))
print(bai1(n))

    
```

```python
#W6A2
def ttl(n):
    A = []
    total = 0
    for i in n:
        total += i
        A.append(total)
    return A
n = list(map(int, input().split()))
print(ttl(n))

    
  
        
    
        
        
```

```python
# W6A3
def xoay(A, k):
    a = list(A)
    for _ in range(k):
        a.append(a.pop(0))
    return tuple(a)
A = tuple(map(int, input().split()))
k = int(input())
print(xoay(A, k))
```

```python
#W6A4
def dic(n):
    d = {}
    for i in n:
        key, value = i.split(":")
        if key not in d:
            d[key] = []
        d[key].append(value)
    return d
    
n = input().split()
print(dic(n))
```

```python
#W6A5
def check(A):
    d = {"positives": 0, "negatives": 0, "zeros": 0}
    for i in A:
         if i > 0:
             d["positives"] += 1
         elif i < 0:
             d["negatives"] += 1
         else:
             d["zeros"] += 1
    return d
A = list(map(int, input().split()))
print(check(A))
    
    
```

```python
#W6A9
def dic(L):
    d = {}
    for i in L:
        key, value = i.split(":")
        d[key] = value
    return d
A = input().split()
B = input().split()
dA = dic(A)
dB = dic(B)
D = {}
for a in dA:
    D[a] = []
    D[a].append(dA[a])
for b in dB:
    if b not in D:
        D[b] = []
    D[b].append(dB[b])
print(dict(sorted(D.items())))
            
            
        

    
```

```python
#W6A10
def bai10(A, k):
    L = []
    for i in range (len(A)):
        for j in range(i+1, len(A)):
            if A[i] + A[j] == k:
                L.append((A[i], A[j]))
    L.sort()
    return L
A = list(map(int, input().split()))
k = int(input())
print(bai10(A, k))
                
        
```

```python
#W6A12
A = list(map(int, input().split()))
d = {}
for i in A:
    if i not in d:
        d[i] = 1
    else:
        d[i] += 1
m = max(d.values())
keys = [k for k, v in d.items() if v == m]
print(min(keys))   
    
    
    
```

```python
#W6A13
keys = input().split()
values = input().split()
d = {}
for k, v in zip(keys, values):
    if v not in d.values():
        d[k] = v
    else:
        print(f"Trùng value {v} nên bỏ qua key {k}")
reverse_d = {v: k for k, v in d.items()}
print(reverse_d)
```

```python
#W6A14
A = list(map(int, input().split()))
B = list(map(int, input().split()))
L = set()
for i in A:
    for j in B:
        if i == j:
            L.add(i)
L = list(L)
print(L)
```
